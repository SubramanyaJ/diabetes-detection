```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Load dataset
file_path = "data.csv"
df = pd.read_csv(file_path)

# Basic statistics
print("Dataset Info:\n", df.info())
print("\nMissing Values:\n", df.isnull().sum())
print("\nSummary Statistics:\n", df.describe())

# Correlation matrix
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Feature Correlation Matrix")
plt.show()

# Diabetes label mapping
diabetes_labels = {0: "Prediabetic", 1: "Type 1 Diabetes", 2: "Type 2 Diabetes", 3: "Type 3c Diabetes"}
df['Diabetes Type'] = df['Target'].map(diabetes_labels)


# Feature distribution across diabetes types
features_to_check = ["Blood Glucose Levels", "Insulin Levels", "Age", "BMI", "Waist Circumference", "Cholesterol Levels", "Pancreatic Health"]
for feature in features_to_check:
    plt.figure(figsize=(8,4))
    sns.boxplot(x='Diabetes Type', y=feature, data=df, order=["Prediabetic", "Type 1 Diabetes", "Type 2 Diabetes", "Type 3c Diabetes"])
    plt.title(f'{feature} Distribution Across Diabetes Types')
    plt.xticks(rotation=45)
    plt.show()


# Check mean values per diabetes type
print("\nMean Feature Values by Diabetes Type:\n", df.groupby('Diabetes Type')[features_to_check].mean().loc[["Prediabetic", "Type 1 Diabetes", "Type 2 Diabetes", "Type 3c Diabetes"]])
```
